```python
# Import library
import pandas as pd
import great_expectations as ge
from great_expectations.data_context import FileDataContext
```

```python
context = FileDataContext.create(project_root_dir='./')
```

```python
# Create data source
datasource = context.sources.add_pandas('main-datasource')

# Create data asset
path = '/content/P2M3_MuhammadFarhan_Hendriyanto_KimiaFarma_Dataset_sampling_clean.csv'
asset = datasource.add_csv_asset('Kimia Farma Sales Analysis', filepath_or_buffer=path)

# Build batch request
batch_request = asset.build_batch_request()
```

```python
# Create an expectation suite
context.add_or_update_expectation_suite('gx_val_suite')

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = 'gx_val_suite'
)

validator.head()
```

```python
# Expectation 1: to be unique
validator.expect_column_values_to_be_unique("id")
```

Jadi di sini kita ingin melakukan validasi apakah kolom ID itu unik? Hal ini dilakukan karena kolom ID merupakan primary key dan harus lah bernilai unik.

```python
# Expectation 2: to be between min_value and max_value
validator.expect_column_values_to_be_between(
    column="rating_cabang", min_value=1, max_value=5
)
```

Jadi di sini saya ingin melakukan validasi untuk memastikan bahwa nilai dari rating_cabang harus memiliki range dari 1-5.

```python
# Expectation 3: to be in set
validator.expect_column_values_to_be_in_set(
    column="branch_category", value_set=["Apotek", "Klinik & Apotek	", "Klinik-Apotek-Laboratorium"]
)
```

Jadi di sini saya ingin melakukan validasi untuk memastikan bahwa nilai yang ada di kategori cabang haruslah paten (seperti set yang immutable) yakni ["Apotek", "Klinik & Apotek	", "Klinik-Apotek-Laboratorium"]

```python
# Expectation 4: to be in type list
validator.expect_column_values_to_be_of_type(
    column="opname_stock", type_="int64"
)
```

Di sini saya ingin memastikan tipe data untuk kolom stok haruslah ber tipe integer. Hal ini bertujuan agar nilai stok selalu tidak menjadi desimal ataupun bertipe string.

```python
# Expectation 5: to be max between
validator.expect_column_max_to_be_between(
    column="rating_transaksi", min_value=1, max_value=5
)
```

Di sini saya ingin memastikan agar nilai rating itu maksimalnya di angka 5

```python
# Expectation 6: column to exist
validator.expect_column_to_exist('transaction_id')
```

Di sini saya ingin melakukan validasi untuk kolom transaksi id di mana kolom tersebut harus ada. Hal ini karena jika tidak ada transaksi id maka tidak terjadi transaksi.

```python
# Expectation 7: column values to not be null
validator.expect_column_values_to_not_be_null("price")
```

Di sini saya ingin memastikan bahwa kolom price tidak boleh ada nilai null. Hal ini karena kolom price nanti akan mempengaruhi kolom net sales. Jadi ini tidak boleh ada nilai kosong.

```python
# Simpan suite
validator.save_expectation_suite(discard_failed_expectations=False)
```

```python
# Check point
checkpoint = context.add_or_update_checkpoint(
    name="gx_val_checkpoint",
    validator=validator
)

checkpoint_result = checkpoint.run()
```
